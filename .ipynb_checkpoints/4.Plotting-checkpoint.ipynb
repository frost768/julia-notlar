# Plotting (Grafikleme,Çizim yapma)
- Standart kütüphane **Plots.jl** olup Pkg.add("Plots") ile yüklenip, **using Plots** komutu ile ortama yüklenebilir.
- **Plots.jl** diğer plotting kütüphaneleri ile iletişim kuran bir üst kütüphanedir.
- **PyPlot, GR, UnicodePlots** (REPL üzerinde grafik için)**, PlotlyJS** gibi plotting kütüphanelerini yine **Pkg.add()** veya **]add lib_1 lib 2 lib3** yükleyip bu taban kütüphaneleri kod içinde ihtiyacınıza göre istediğiniz zaman **lib_adi();** ile çağırarak kullanabilrisiniz
    - PlotlyJS, **ORCA** adında başka bir kütüphane gerektiryor bu yüzden PlotlyJS kullanmak için:
        - Pkg.add("ORCA")
        - Pkg.add("PlotlyJS")
    - Sırasıyla eklemeniz önerilir.
- **plot(x,y,...)** : 
    - x,y sonrası daha çok özelleştirme için argümanlar var.
    - current() objesi plot komutuyla oluşur ve grafiğinizi temsil eder. Grafiği tekrar görmek için current() fonksiyonunu kullanabilirsiniz.
- **plot!(...)** : 
    - Önceden de söylendiği gibi **!** işareti bir değişkeni değiştirdiğimizi belirtiyordu. Bu komut önceden oluşturulmuş bir grafiğe yeni çizimler eklemeyi sağlar yani aynı grafik üzerinde farklı çizimler gösterebilirsiniz.
    - Birinci argüman olarak bir plot objesi alırsa o grafiği değiştirir. Eğer verilmemişse current() varsayılanını değiştirir.
    - Matrislerin her bir sütunu bir seri olarak alınıp y değeri gibi kullanılır.  
    - Grafikleri kaydetmek için
        - savefig(plot_objesi, "dosyaadı.uzantı") olarak dışarı aktarabilirsiniz  
        
# Özelleştirme argümanlarından birkaçı
- **xlabel:** x ekseni adı
- **ylabel:** y ekseni adı
- **label:** noktaları legend (grafik açıklaması) üzerinde görünecek olan ve noktaların neyi belirttiğini görmek için kullanabilirsiniz.
- **title:** grafik başlığı
- **size:** grafik boyutu size=(x,y)
- **linewidth/lw:** Çizgi kalınlığı
- **layout:** Birden fazla seriyi farklı grafiklerde ama aynı tablo/kutu içinde gösterebiliriz.
    - plot(seri1,seri2,...,layout=(2,1) : 2 satırlı ve 1 sütünlu tablo/kutu içerisinde *seri1 ve seri2'yi* grafikle
- **seriestype:** serilerin nasıl grafik üzerinde gösterileceğini değiştiren komut
    - *histogram, scatter,bar,line...*
    - seri tipleri :seritipi_adi ile kullanılır. örneğin **plot(..., seriestype= :scatter,...)**
    - bunun dışında hepsi yine kendi fonksiyonuna sahip olup **seritipi_adi(plot_objesi)** ile kullabılabilir
- Diğer argümanları ve hangi kütüphane tarafından desteklendiğini görmek için https://docs.juliaplots.org/latest/generated/supported/#supported-1 linkine göz atabilrisiniz.

```julia
using Plots
gr();
```

```julia
#Basit bir 2.derece fonksiyonu çizdirelim
f(x)=x^2
x=collect(-2:0.1:2)
y=f.(x)
plot(x,y)
```

```julia
plot!(
    0:0.1:3, # x
    f.(0:0.1:3), # y
    xlabel="x değerleri",
    ylabel="x^2 değerleri",
    title="[0,2] ve [0,3] aralıklarının karşılaştırılması",
    lw=3
)
```

```julia
# fonksiyonun geçtiği noktaları belirtelim
scatter!(x,y,label="f(x)=x^2")
```

```julia
## https://docs.juliaplots.org/latest/tutorial/ sitesi örneği
p1 = plot(x, y) # Çizgi grafiği
p2 = scatter(x, y) # dağılım (scatter) grafiği
p3 = bar(x, y, title = "Bar grafiği", lw = 3)
p4 = histogram(x, y) # histogram
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```

```julia
# Daha düzenli bir çalışma için
x = collect(0:0.05:1.2)
y = [sin.(x) cos.(x) exp.(2x) x.^x ]
basliklar = ["sin(x)" "cos(x)" "ℯ^2x" "x^x"]
renkler = [ :red :orange :blue :fuchsia ]
line=plot(
    x,
    y,
    linestyle=:dot,
    arrow=:arrow,
    title=basliklar,
    color=renkler,
    layout=(4,1),
    linewidth=3,
    legend=false
)

dagitim=scatter(
    x,
    y,
    title=basliklar,
    color=renkler,
    layout=(2,2),
    legend=false
)
plot(line,dagitim,bg_color=:grey)
```

```julia
# 3D Plot
x=0:pi
y=sin.(x).*cos.(x)
z=x.*y
plot3d(
    x,y,z,
    line=(:red,3,:dot),
    title="x ve y değerlerinin 3.boyuta yansıması"
        )
```
