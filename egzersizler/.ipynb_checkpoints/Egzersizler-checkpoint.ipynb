```julia
using Plots; gr();
function checkfermat(a,b,c,n)
    return n>2 && a^n + b^n ==c^n ? println("Fermat is wrong") : println("Fermat is right")
end
checkfermat(1,23,5,3)
```

```julia
#Ackermann ack(3,4)=125, m>n için stackoverflow
function ack(m,n)
    if(n<0 || m<0)
        println("m veya n < 0 olamaz")
        return -1
    elseif(m==0)
        return n+1
    elseif(n==0)
        return ack(m-1,1)
    else
        return ack(m-1,ack(m,n-1))
    end
end
plot([ack(2,i+1) for i in 1:10],title="Ackermann")
```

```julia
#COLLATZ CONJECTURE
function collatz(n)
    while n !=1
    if(n%2==0)
        return n/2
    else
        return 3n+1
    end
    end
end
p1=histogram(collatz.(2:100))
p2=bar(collatz.(2:100))
plot(p1,p2,layout=(1,2))
```

```julia
# REPETITION İLE KÖK BULMA
kok_bul(a,x)=(1/2)*(x+a/x)
tahminler=[]
farklar=[Inf]
function karsilastir(a,x)
    if(a<0 || x <0)
        print("A ve X <0 olamaz")
        return -1
    end
    tahmin=kok_bul(a,x)
    push!(tahminler,tahmin)
    gercek=sqrt(a)
    println("tahmin" *'\t'^3* "gercek" *'\t'* "fark")
    while true
        tahmin=kok_bul(a,tahmin)
        push!(tahminler,tahmin)
        fark=abs(tahmin-gercek)
        push!(farklar,fark)
        println("$tahmin" *" "^6* "$gercek" *" "^6 * "$fark")
        if(fark<10^-15)
            break
        end
    end
end
karsilastir(14350,1)
p1=plot(tahminler,title="Tahminler")
p2=plot(farklar,title="Farklar")
plot(p1,p2,layout=(2,1),legend=false)
```

```julia
# KARAKTER BUL
function bul(kelime::String,karakter::Char)
    if(isempty(kelime)|| isempty(karakter))
        return 0
    end
    i=firstindex(kelime)
    while i<=sizeof(kelime)
        if(karakter==kelime[i])
            return i
        end
        i=nextind(kelime,i)
    end
    -1
end
# METHOD 2
function bul(kelime::AbstractString,karakter::Char,baslangic_i::Int)
    if(isempty(kelime)|| isempty(karakter)|| baslangic_i<1)
        return 0
    end
    i=baslangic_i
    while i<=sizeof(kelime)
        if(karakter==kelime[i])
            return i
        end
        i=nextind(kelime,i)
    end
    -1
end
methods(bul) # 3 tane
bul("abc",'a') # 1
 #bul("abc",'a',0) # 0
```

```julia
# KARAKTER SAY
function say(kelime,karakter)
    sayac=0
    for k in kelime
        if(k==karakter)
            sayac+=1
        end
    end
    return sayac
end
say("atatürk",'a')

function say2(kelime,k)
    i=1
    sayac=0
    while bul(kelime,k,i)>0
        sayac+=1
        i=bul(kelime,k,i)+1
    end
    return sayac
end
say2("atatürk",'a')
```

```julia
# ROT3
function ceaser_cypher(str,r)
    str=lowercase(str)
    cyphered=[]
    for i in str
       @show  c=Int(i)+r%26
        if(c<123)
            push!(cyphered, Char(c%123))
        else
             push!(cyphered, Char(c%97)+71)
        end
    end
    return cyphered
end
ceaser_cypher("cheer",7)
```

```julia
# Nested array sum
function toplaArrOfArr(arr)
    return sum(map(x-> sum(x), arr))
end
a=[vec(rand(4)) for _ in 1:5]

topla(a)
```

```julia
# kümülatif toplam
function cumulsum(arr)
    new_arr=[arr[1]]
    for i in 1:length(arr)-1
        push!(new_arr, new_arr[i]+arr[i+1])
    end
    new_arr
end
cumulsum(collect(1:5))
```

```julia
# interior
function interior!(arr)
    if length(arr)<3
        print("Hata")
    else
        deleteat!(arr,1)
        deleteat!(arr,lastindex(arr))
    end
    nothing
end
interior!(collect(1:5))
```

```julia
# is sorted
function issort(arr)
    return arr==sort(arr)
end
issort([3,4,3,4])
issort(collect(1:6))
```

```julia
# is anagram
function isanagram(str1,str2)
    return sort(collect(str1))==sort(collect(str2))
end
isanagram("balm","lamb")
```

```julia
# has duplicates
function hasduplicates(arr)::Bool
    return !(intersect(arr,arr)==arr) #bir dizinin kendisi ile kesişimi o dizinin unique elemanlarını verir
end
hasduplicates([4,3,2,1,1])
```
